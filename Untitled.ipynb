```python
import numpy as np
from numpy import cos,sin,pi
import matplotlib.pyplot as plt
```

```python
f = 1.
w = 2*pi*f
Es = 1.
Zln = 1.
Zld = 1.
theta = pi/3
phi = pi/2

def f(t,w,Es,Zln,Zld,theta,phi):
    a = Zln*cos(theta)+Zld*cos(phi)
    b = Zln*sin(theta)+Zld*sin(phi)
    return Es*(cos(w*t)*a + np.sin(w*t)*b) / (a**2 + b**2)

def g(t,w,Es,Zln,Zld,theta,phi):
    a = Zln*cos(theta)+Zld*cos(phi)
    b = Zln*sin(theta)+Zld*sin(phi)
    return Es*cos(w*t) / np.sqrt(a**2 + b**2)

t = np.linspace(0,1,100)
x = f(t,w,Es,Zln,Zld,theta,phi)
y = g(t,w,Es,Zln,Zld,theta,phi)

plt.plot(t,x,'k')
plt.plot(t,y,'r')
plt.show()
```
